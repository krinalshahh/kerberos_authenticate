```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse

def is_valid_url(url):
    if url is None:
        return False
    url = url.strip()
    if url == '':
        return False
    if url.startswith('#'):
        return False
    if url.lower().startswith('javascript'):
        return False
    parsed = urlparse(url)
    # Only allow http or https URLs or relative URLs without scheme
    if parsed.scheme and parsed.scheme not in ['http', 'https']:
        return False
    return True

def crawl(url, max_depth=3, current_depth=0, visited=None):
    if visited is None:
        visited = set()
    if current_depth > max_depth:
        return visited
    if url in visited:
        return visited

    print(f"Crawling ({current_depth}): {url}")
    visited.add(url)

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
    except Exception as e:
        print(f"Failed to fetch {url}: {e}")
        return visited

    soup = BeautifulSoup(response.content, "html.parser")

    for a in soup.find_all("a", href=True):
        href = a['href'].strip()
        if not is_valid_url(href):
            continue
        full_url = urljoin(url, href)
        # Optionally: restrict to same domain to avoid crawling the whole web
        if urlparse(full_url).netloc != urlparse(url).netloc:
            continue
        if full_url not in visited:
            crawl(full_url, max_depth, current_depth + 1, visited)

    return visited

if __name__ == "__main__":
    start_url = "https://economictimes.indiatimes.com/news"
    all_urls = crawl(start_url, max_depth=3)

    print("\nAll crawled URLs:")
    for u in all_urls:
        print(u)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time
from fake_useragent import UserAgent

class NPRNewsCrawler:
    def __init__(self, max_articles=5, delay=2):
        self.base_url = "https://www.npr.org/sections/news/"
        self.max_articles = max_articles
        self.delay = delay
        self.visited_urls = set()
        self.articles = []
        self.ua = UserAgent()
        self.session = requests.Session()
        self.session.headers.update({
            'User-Agent': self.ua.random,
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Referer': 'https://www.google.com/',
        })

    def _is_article_url(self, url):
        # NPR article URLs typically contain numbers (e.g., /2023/12/15/1214567890/title)
        return url.startswith('https://www.npr.org/') and any(char.isdigit() for char in url) and url not in self.visited_urls

    def _extract_article(self, url):
        try:
            time.sleep(self.delay)
            response = self.session.get(url)
            response.raise_for_status()

            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract title
            title = soup.find('h1').get_text(strip=True) if soup.find('h1') else "No Title"

            # Extract content - NPR uses <div id="storytext">
            body = soup.find('div', id='storytext') or soup.find('article')
            content = body.get_text(separator='\n', strip=True) if body else "No Content"

            return {
                'url': url,
                'title': title,
                'content': content[:2000] + "..." if len(content) > 2000 else content,
                'timestamp': soup.find('time')['datetime'] if soup.find('time') else None
            }
        except Exception as e:
            print(f"Error processing {url}: {str(e)}")
            return None

    def _get_article_links(self, url):
        try:
            time.sleep(self.delay)
            response = self.session.get(url)
            response.raise_for_status()

            soup = BeautifulSoup(response.text, 'html.parser')
            links = set()

            # Find article links in NPR's structure
            for a in soup.select('a[href^="/"]'):
                href = a['href']
                if any(char.isdigit() for char in href):  # URLs with numbers are likely articles
                    absolute_url = urljoin(self.base_url, href)
                    if self._is_article_url(absolute_url):
                        links.add(absolute_url)

            return links
        except Exception as e:
            print(f"Error fetching links from {url}: {str(e)}")
            return set()

    def crawl(self):
        print(f"Starting crawl on {self.base_url}...")
        article_links = self._get_article_links(self.base_url)

        for link in list(article_links)[:self.max_articles]:
            if len(self.articles) >= self.max_articles:
                break

            print(f"Processing: {link}")
            article = self._extract_article(link)
            if article:
                self.articles.append(article)
                self.visited_urls.add(link)
                print(f"✓ Article saved ({len(self.articles)}/{self.max_articles})")

        return self.articles

    def save_to_file(self, filename="npr_articles.txt"):
        with open(filename, 'w', encoding='utf-8') as f:
            for article in self.articles:
                f.write(f"URL: {article['url']}\n")
                f.write(f"Title: {article['title']}\n")
                f.write(f"Date: {article.get('timestamp', 'Unknown')}\n")
                f.write(f"Content:\n{article['content']}\n")
                f.write("\n" + "="*80 + "\n\n")
        print(f"Saved {len(self.articles)} articles to {filename}")

if __name__ == "__main__":
    # Initialize crawler (max 5 articles, 2 sec delay between requests)
    crawler = NPRNewsCrawler(max_articles=100, delay=2)

    # Start crawling
    articles = crawler.crawl()

    # Display results
    print("\nCrawling complete. Results:")
    for i, article in enumerate(articles, 1):
        print(f"\nArticle {i}: {article['title']}")
        print(f"Date: {article.get('timestamp', 'Unknown')}")
        print(f"URL: {article['url']}")
        print(f"Preview: {article['content'][:200]}...")

    # Save to file
    crawler.save_to_file()
```

```python
# Nested News Scraper with Date Filter
# Install required packages first: pip install requests beautifulsoup4 pandas openpyxl python-dateutil

import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
from datetime import datetime, timedelta
from dateutil import parser
import re
import time
from urllib.parse import urljoin, urlparse
import logging
from typing import Dict, List, Optional
import warnings
warnings.filterwarnings('ignore')

# Setup logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

class NestedNewsScraper:
    def __init__(self, delay=1, timeout=10):
        self.session = requests.Session()
        self.session.headers.update({
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        })
        self.delay = delay
        self.timeout = timeout
        self.scraped_data = {}

    def get_page(self, url: str) -> Optional[BeautifulSoup]:
        """Fetch and parse a webpage"""
        try:
            response = self.session.get(url, timeout=self.timeout)
            response.raise_for_status()
            return BeautifulSoup(response.content, 'html.parser')
        except Exception as e:
            logger.error(f"Error fetching {url}: {str(e)}")
            return None

    def extract_date(self, text: str) -> Optional[datetime]:
        """Extract date from text using various formats"""
        if not text:
            return None

        # Clean the text
        text = re.sub(r'[^\w\s\-/:,.]', '', text.strip())

        # Common date patterns
        date_patterns = [
            r'\d{4}-\d{2}-\d{2}',  # 2024-01-15
            r'\d{2}/\d{2}/\d{4}',  # 01/15/2024
            r'\d{2}-\d{2}-\d{4}',  # 01-15-2024
            r'\w+ \d{1,2}, \d{4}', # January 15, 2024
            r'\d{1,2} \w+ \d{4}',  # 15 January 2024
        ]

        for pattern in date_patterns:
            match = re.search(pattern, text)
            if match:
                try:
                    return parser.parse(match.group())
                except:
                    continue

        # Try parsing the entire text
        try:
            return parser.parse(text)
        except:
            return None

    def is_target_date(self, article_date: datetime, target_date: datetime) -> bool:
        """Check if article date matches target date"""
        if not article_date:
            return False
        return article_date.date() == target_date.date()

    def find_category_links(self, soup: BeautifulSoup, base_url: str, categories: List[str]) -> Dict[str, str]:
        """Find links for specified categories"""
        category_links = {}

        # Special handling for Hacker News
        if 'ycombinator.com' in base_url:
            hn_categories = {
                'newest': 'https://news.ycombinator.com/newest',
                'show': 'https://news.ycombinator.com/show',
                'ask': 'https://news.ycombinator.com/ask',
                'jobs': 'https://news.ycombinator.com/jobs',
                'best': 'https://news.ycombinator.com/best',
                'front': 'https://news.ycombinator.com/'
            }

            for category in categories:
                if category.lower() in hn_categories:
                    category_links[category] = hn_categories[category.lower()]

            return category_links

        # Common selectors for navigation links
        nav_selectors = [
            'nav a', '.nav a', '.menu a', '.navigation a',
            '.category-links a', '.section-links a',
            'header a', '.header a', '.main-nav a'
        ]

        all_links = []
        for selector in nav_selectors:
            links = soup.select(selector)
            all_links.extend(links)

        # Also check for any links in the page
        all_links.extend(soup.find_all('a', href=True))

        for category in categories:
            category_lower = category.lower()
            for link in all_links:
                href = link.get('href', '')
                text = link.get_text(strip=True).lower()

                if category_lower in text or category_lower in href.lower():
                    full_url = urljoin(base_url, href)
                    category_links[category] = full_url
                    break

        return category_links

    def extract_news_articles(self, soup: BeautifulSoup, base_url: str, target_date: datetime, max_articles: int = 20) -> List[Dict]:
        """Extract news articles from a page"""
        articles = []

        # Check if this is Hacker News
        if 'ycombinator.com' in base_url:
            return self.extract_hackernews_articles(soup, base_url, target_date, max_articles)

        # Common selectors for news articles
        article_selectors = [
            'article', '.article', '.news-item', '.post',
            '.story', '.entry', '.content-item'
        ]

        # Common selectors for headlines
        title_selectors = [
            'h1', 'h2', 'h3', '.headline', '.title', '.news-title',
            '.post-title', '.entry-title', '.story-title'
        ]

        # Common selectors for dates
        date_selectors = [
            '.date', '.publish-date', '.timestamp', 'time',
            '.post-date', '.entry-date', '.news-date'
        ]

        # Try to find article containers first
        article_elements = []
        for selector in article_selectors:
            elements = soup.select(selector)
            if elements:
                article_elements = elements[:max_articles * 2]
                break

        # If no article containers found, look for title elements
        if not article_elements:
            for selector in title_selectors:
                elements = soup.select(selector)
                if elements:
                    article_elements = elements[:max_articles * 2]
                    break

        # If still no elements, try more generic approach
        if not article_elements:
            # Look for any links that might be articles
            links = soup.find_all('a', href=True)
            article_elements = links[:max_articles * 2]

        for element in article_elements:
            if len(articles) >= max_articles:
                break

            article_data = {}

            # Extract title
            title_text = ""
            if element.name == 'a':
                title_text = element.get_text(strip=True)
            else:
                title_elem = element if element.name in ['h1', 'h2', 'h3'] else None
                if not title_elem:
                    for selector in title_selectors:
                        title_elem = element.select_one(selector)
                        if title_elem:
                            break

                if title_elem:
                    title_text = title_elem.get_text(strip=True)

            if not title_text or len(title_text) < 5:
                continue

            article_data['title'] = title_text

            # For date filtering, if no specific date found, include recent articles
            # This is more permissive for testing
            article_data['date'] = target_date.strftime('%Y-%m-%d')
            article_data['date_text'] = 'Estimated current date'

            # Try to find actual date
            date_found = False
            for selector in date_selectors:
                date_elem = element.select_one(selector) if hasattr(element, 'select_one') else None
                if date_elem:
                    date_text = date_elem.get_text(strip=True)
                    if not date_text and date_elem.get('datetime'):
                        date_text = date_elem.get('datetime')

                    article_date = self.extract_date(date_text)
                    if article_date:
                        article_data['date'] = article_date.strftime('%Y-%m-%d')
                        article_data['date_text'] = date_text
                        date_found = True
                        break

            # Extract link
            if element.name == 'a' and element.get('href'):
                article_data['link'] = urljoin(base_url, element['href'])
            else:
                link_elem = element.find('a', href=True) if hasattr(element, 'find') else None
                if link_elem:
                    article_data['link'] = urljoin(base_url, link_elem['href'])

            article_data['scraped_at'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            articles.append(article_data)

        return articles

    def extract_hackernews_articles(self, soup: BeautifulSoup, base_url: str, target_date: datetime, max_articles: int = 20) -> List[Dict]:
        """Specialized extraction for Hacker News"""
        articles = []

        # Hacker News specific selectors
        story_rows = soup.select('tr.athing')

        for i, row in enumerate(story_rows[:max_articles]):
            if len(articles) >= max_articles:
                break

            article_data = {}

            # Extract title and link
            title_link = row.select_one('span.titleline > a')
            if title_link:
                article_data['title'] = title_link.get_text(strip=True)
                href = title_link.get('href', '')
                if href.startswith('item?'):
                    article_data['link'] = urljoin(base_url, href)
                else:
                    article_data['link'] = href
            else:
                continue

            # Get the next row which contains metadata
            next_row = row.find_next_sibling('tr')
            if next_row:
                # Extract score
                score_elem = next_row.select_one('span.score')
                if score_elem:
                    article_data['score'] = score_elem.get_text(strip=True)

                # Extract user and time
                user_elem = next_row.select_one('a.hnuser')
                if user_elem:
                    article_data['author'] = user_elem.get_text(strip=True)

                time_elem = next_row.select_one('span.age')
                if time_elem:
                    time_text = time_elem.get_text(strip=True)
                    article_data['time_ago'] = time_text

                    # For HN, we'll include articles from today (most recent)
                    # Since HN doesn't show exact dates, we'll be permissive
                    if any(term in time_text.lower() for term in ['hour', 'minute', 'day']):
                        article_data['date'] = target_date.strftime('%Y-%m-%d')
                        article_data['date_text'] = time_text
                    else:
                        continue  # Skip older articles

                # Extract comments count
                comments_elem = next_row.select_one('a[href*="item?id="]')
                if comments_elem and 'comment' in comments_elem.get_text().lower():
                    article_data['comments'] = comments_elem.get_text(strip=True)

            article_data['scraped_at'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
            article_data['source'] = 'Hacker News'
            articles.append(article_data)

        return articles

    def scrape_nested_structure(self, main_url: str, categories: List[str], target_date: datetime, max_depth: int = 3, max_articles: int = 20):
        """Main scraping function with nested structure"""
        logger.info(f"Starting nested scraping for {main_url}")
        logger.info(f"Target date: {target_date.strftime('%Y-%m-%d')}")
        logger.info(f"Categories: {categories}")

        self.scraped_data = {
            'main_url': main_url,
            'target_date': target_date.strftime('%Y-%m-%d'),
            'scraped_at': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'categories': {}
        }

        # Get main page
        main_soup = self.get_page(main_url)
        if not main_soup:
            logger.error("Failed to fetch main page")
            return

        # Find category links
        category_links = self.find_category_links(main_soup, main_url, categories)
        logger.info(f"Found category links: {category_links}")

        # Scrape each category
        for category, category_url in category_links.items():
            logger.info(f"Scraping category: {category} - {category_url}")

            category_soup = self.get_page(category_url)
            if not category_soup:
                logger.warning(f"Failed to fetch category page: {category}")
                continue

            # Extract articles from category page
            articles = self.extract_news_articles(category_soup, category_url, target_date, max_articles)

            self.scraped_data['categories'][category] = {
                'url': category_url,
                'articles_count': len(articles),
                'articles': articles
            }

            logger.info(f"Found {len(articles)} articles for {category}")
            time.sleep(self.delay)  # Rate limiting

        # If no category links found, try to extract from main page
        if not category_links:
            logger.info("No category links found, extracting from main page")
            articles = self.extract_news_articles(main_soup, main_url, target_date, max_articles)

            self.scraped_data['categories']['general'] = {
                'url': main_url,
                'articles_count': len(articles),
                'articles': articles
            }

        total_articles = sum(cat['articles_count'] for cat in self.scraped_data['categories'].values())
        logger.info(f"Scraping completed. Total articles found: {total_articles}")

    def export_to_json(self, filename: str = None):
        """Export scraped data to JSON"""
        if not filename:
            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
            filename = f'news_data_{timestamp}.json'

        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(self.scraped_data, f, indent=2, ensure_ascii=False)

        logger.info(f"Data exported to {filename}")
        return filename

    def export_to_excel(self, filename: str = None):
        """Export scraped data to Excel"""
        if not filename:
            timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
            filename = f'news_data_{timestamp}.xlsx'

        # Flatten data for Excel
        rows = []
        for category, data in self.scraped_data['categories'].items():
            for article in data['articles']:
                row = {
                    'category': category,
                    'category_url': data['url'],
                    'title': article.get('title', ''),
                    'date': article.get('date', ''),
                    'link': article.get('link', ''),
                    'summary': article.get('summary', ''),
                    'scraped_at': article.get('scraped_at', '')
                }
                rows.append(row)

        df = pd.DataFrame(rows)

        # Create Excel file with multiple sheets
        with pd.ExcelWriter(filename, engine='openpyxl') as writer:
            # Main sheet with all data
            df.to_excel(writer, sheet_name='All_News', index=False)

            # Separate sheet for each category
            for category in self.scraped_data['categories'].keys():
                category_df = df[df['category'] == category]
                if not category_df.empty:
                    sheet_name = category.replace(' ', '_')[:31]  # Excel sheet name limit
                    category_df.to_excel(writer, sheet_name=sheet_name, index=False)

            # Summary sheet
            summary_data = []
            for category, data in self.scraped_data['categories'].items():
                summary_data.append({
                    'Category': category,
                    'URL': data['url'],
                    'Articles_Count': data['articles_count']
                })

            summary_df = pd.DataFrame(summary_data)
            summary_df.to_excel(writer, sheet_name='Summary', index=False)

        logger.info(f"Data exported to {filename}")
        return filename

    def get_stats(self):
        """Get scraping statistics"""
        if not self.scraped_data:
            return "No data scraped yet"

        total_articles = sum(cat['articles_count'] for cat in self.scraped_data['categories'].values())
        categories_count = len(self.scraped_data['categories'])

        stats = f"""
Scraping Statistics:
- Main URL: {self.scraped_data['main_url']}
- Target Date: {self.scraped_data['target_date']}
- Categories Found: {categories_count}
- Total Articles: {total_articles}
- Scraped At: {self.scraped_data['scraped_at']}

Categories Breakdown:
"""

        for category, data in self.scraped_data['categories'].items():
            stats += f"- {category}: {data['articles_count']} articles\n"

        return stats

# Example usage and configuration
def main():
    # Configuration - Modify these values as needed
    MAIN_URL = "https://example-news-site.com"  # Replace with actual news site
    TARGET_DATE = datetime.now()  # Today's date - user can change this
    CATEGORIES = ["world", "japan", "politics", "sports", "technology"]  # Modify as needed
    MAX_DEPTH = 3
    MAX_ARTICLES_PER_CATEGORY = 20

    # Alternative: Set specific date
    # TARGET_DATE = datetime(2024, 5, 26)  # Specific date

    # Initialize scraper
    scraper = NestedNewsScraper(delay=1, timeout=10)

    print("🚀 Starting Nested News Scraper")
    print(f"📅 Target Date: {TARGET_DATE.strftime('%Y-%m-%d')}")
    print(f"🌐 Main URL: {MAIN_URL}")
    print(f"📂 Categories: {', '.join(CATEGORIES)}")
    print("-" * 50)

    # Start scraping
    scraper.scrape_nested_structure(
        main_url=MAIN_URL,
        categories=CATEGORIES,
        target_date=TARGET_DATE,
        max_depth=MAX_DEPTH,
        max_articles=MAX_ARTICLES_PER_CATEGORY
    )

    # Display statistics
    print(scraper.get_stats())

    # Export data
    json_file = scraper.export_to_json()
    excel_file = scraper.export_to_excel()

    print(f"\n✅ Export completed:")
    print(f"📄 JSON: {json_file}")
    print(f"📊 Excel: {excel_file}")

    # Display sample data
    if scraper.scraped_data['categories']:
        print("\n📰 Sample Articles:")
        for category, data in list(scraper.scraped_data['categories'].items())[:2]:
            print(f"\n{category.upper()}:")
            for article in data['articles'][:3]:  # Show first 3 articles
                print(f"  - {article.get('title', 'No title')}")
                print(f"    Date: {article.get('date', 'No date')}")
                print(f"    Link: {article.get('link', 'No link')}")

    return scraper

# Run the scraper
if __name__ == "__main__":
    # You can modify these parameters before running
    scraper = main()

    # Additional usage examples:

    # Change target date to yesterday
    # yesterday = datetime.now() - timedelta(days=1)
    # scraper.scrape_nested_structure("https://news-site.com", ["world", "tech"], yesterday)

    # Scrape specific categories only
    # scraper.scrape_nested_structure("https://news-site.com", ["japan"], datetime.now(), max_articles=10)

    # Export with custom filename
    # scraper.export_to_json("my_news_data.json")
    # scraper.export_to_excel("my_news_data.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
import time
from fake_useragent import UserAgent
from collections import deque

class NewsCrawler:
    def __init__(self, start_url, max_articles=20, max_depth=5, delay=1.5):
        self.start_url = start_url
        self.max_articles = max_articles
        self.max_depth = max_depth
        self.delay = delay
        self.visited_urls = set()
        self.articles = []
        self.ua = UserAgent()
        self.session = requests.Session()
        self.domain = urlparse(start_url).netloc
        self.queue = deque()

        # Configure session headers
        self.session.headers.update({
            'User-Agent': self.ua.random,
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Referer': 'https://www.google.com/',
            'DNT': '1',
        })

    def _is_valid_url(self, url):
        """Check if URL belongs to the same domain and hasn't been visited"""
        parsed = urlparse(url)
        return (parsed.netloc == self.domain
                and url not in self.visited_urls
                and not any(ext in url.lower() for ext in ['.pdf', '.jpg', '.png']))

    def _is_article_page(self, url, soup):
        """Determine if page is an article (customize for target site)"""
        # Common indicators of article pages
        article = soup.find('article')
        headline = soup.find('h1')
        date_published = soup.find('time') or soup.find('meta', property='article:published_time')

        return (article is not None
                or headline is not None
                or date_published is not None
                or '/article/' in url
                or any(seg in url for seg in ['/news/', '/story/', '/post/']))

    def _extract_article_content(self, soup):
        """Extract main article content (customize for target site)"""
        # Try common article containers
        article = (soup.find('article') or
                  soup.find('div', class_=lambda x: x and 'article' in x.lower()) or
                  soup.find('div', id=lambda x: x and 'content' in x.lower()) or
                  soup.find('main'))

        if not article:
            return None

        # Remove unwanted elements
        for element in article.find_all(['script', 'style', 'nav', 'footer', 'aside', 'figure']):
            element.decompose()

        return article.get_text(separator='\n', strip=True)

    def _process_page(self, url, depth):
        """Process a single page and extract links/articles"""
        try:
            time.sleep(self.delay)
            response = self.session.get(url)
            response.raise_for_status()

            soup = BeautifulSoup(response.text, 'html.parser')
            self.visited_urls.add(url)

            # Check if this is an article page
            if self._is_article_page(url, soup):
                content = self._extract_article_content(soup)
                if content:
                    title = soup.find('h1').get_text(strip=True) if soup.find('h1') else "No Title"
                    date = (soup.find('time') or
                           soup.find('meta', property='article:published_time') or
                           soup.find('span', class_=lambda x: x and 'date' in x.lower()))
                    date = date.get('datetime') if hasattr(date, 'get') else date.get_text(strip=True) if date else "Unknown"

                    self.articles.append({
                        'url': url,
                        'title': title,
                        'date': date,
                        'content': content[:5000] + "..." if len(content) > 5000 else content,
                        'depth': depth
                    })
                    print(f"✓ Article found at depth {depth}: {title[:50]}...")

            # If we haven't reached max depth or max articles, find links to crawl
            if depth < self.max_depth and len(self.articles) < self.max_articles:
                for link in soup.find_all('a', href=True):
                    absolute_url = urljoin(self.start_url, link['href'])
                    if self._is_valid_url(absolute_url):
                        self.queue.append((absolute_url, depth + 1))

        except Exception as e:
            print(f"Error processing {url}: {str(e)}")

    def crawl(self):
        """Start crawling from the initial URL"""
        print(f"\nStarting crawl on {self.start_url} (max depth: {self.max_depth})")
        self.queue.append((self.start_url, 0))

        while self.queue and len(self.articles) < self.max_articles:
            url, depth = self.queue.popleft()
            if url not in self.visited_urls:
                self._process_page(url, depth)

        print(f"\nCrawling complete. Found {len(self.articles)} articles.")

    def save_results(self, filename="news_articles.json"):
        """Save results to a JSON file"""
        import json
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(self.articles, f, ensure_ascii=False, indent=2)
        print(f"Results saved to {filename}")

    def print_summary(self):
        """Print a summary of collected articles"""
        print("\nCollected Articles Summary:")
        for i, article in enumerate(self.articles, 1):
            print(f"\n{i}. {article['title']}")
            print(f"   Depth: {article['depth']} | Date: {article['date']}")
            print(f"   URL: {article['url']}")
            print(f"   Preview: {article['content'][:100]}...")

# Example usage with crawler-friendly sites
if __name__ == "__main__":
    # Choose a crawler-friendly site (tested options)
    sites = {
        'AP News': 'https://apnews.com',
        'PBS NewsHour': 'https://www.pbs.org/newshour/',
        'BBC News': 'https://www.bbc.com/news',
        'NPR News': 'https://www.npr.org/sections/news/'
    }

    print("Available crawler-friendly news sites:")
    for i, (name, url) in enumerate(sites.items(), 1):
        print(f"{i}. {name} ({url})")

    choice = int(input("\nSelect a site to crawl (1-4): ")) - 1
    selected_url = list(sites.values())[choice]

    # Initialize and run crawler
    crawler = NewsCrawler(
        start_url=selected_url,
        max_articles=100,
        max_depth=5,
        delay=1.5
    )
    crawler.crawl()
    crawler.print_summary()
    crawler.save_results()
```
